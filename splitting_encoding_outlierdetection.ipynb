```python
import pandas as pd
import numpy as np
from google.colab import drive
```

```python
file_path = '/content/drive/MyDrive/Protein Sequencing/Protein Sequence Dataset.csv'
df = pd.read_csv(file_path)
```

```python
df
```

```python
df[['gene', 'species']] = df['name'].str.split('_', n=1, expand=True)
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
```

```python
df['EC_number_encoded'] = label_encoder.fit_transform(df['EC_number'])
df
```

```python
EC_number_label_mapping = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))
EC_number_label_mapping
```

```python
df['species_encoded'] = label_encoder.fit_transform(df['species'])
df
```

```python
species_label_mapping = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))
species_label_mapping
```

```python
unique_species = df['species'].unique()
unique_species
```

```python
len(unique_species)
```

```python
df['sequence_len'] = df['sequence'].apply(lambda x: len(x))
df
```

```python
df[df['sequence_len'] == 0]
```

```python
import matplotlib.pyplot as plt
```

```python
filtered_df = df[df['EC_number_encoded'] == 0]

plt.figure(figsize=(20, 10))
plt.plot(filtered_df.index, filtered_df['sequence_len'], linestyle='-')

plt.xlabel('Index')
plt.ylabel('Sequence Length')
plt.title('Sequence Length for EC_number_encoded=0')

plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6, 10))
plt.boxplot(df['sequence_len'])

plt.xlabel('Sequence Length')
plt.ylabel('Value')
plt.title('Boxplot of Sequence Length')

plt.show()
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(200, 50))
sns.boxplot(x='EC_number_encoded', y='sequence_len', data=df)

plt.xlabel('EC_number_encoded')
plt.ylabel('Sequence Length')
plt.title('Boxplot of Sequence Length for Each EC_number_encoded')

plt.xticks(rotation=45)

plt.grid(True)
plt.show()
```
