```python
import pandas as pd
import numpy as np
from google.colab import drive
```

```python
file_path = '/content/drive/MyDrive/Protein Sequencing/Encoded_dataset.csv'
df = pd.read_csv(file_path)
```

```python
df
```

```python
valid_aa_codes = set('ARNDCQEGHILKMFPSTWYV')

# Create a function to check if a sequence is valid
def is_valid_sequence(sequence):
    return all(aa in valid_aa_codes for aa in sequence)

# Apply the function to the 'sequence' column
valid_rows = df['sequence'].apply(is_valid_sequence)

# Filter the dataframe to keep only valid rows
df = df[valid_rows]
```

```python
df
```

```python
amino_acids = 'ACDEFGHIKLMNPQRSTVWY'

# Function to one-hot encode a protein sequence
def one_hot_encode(sequence):
    # Create an empty array to store the one-hot encoded sequence
    encoded_sequence = np.zeros((len(sequence), len(amino_acids)))

    # Iterate through the sequence and set the corresponding position to 1
    for i, aa in enumerate(sequence):
        if aa in amino_acids:
            index = amino_acids.index(aa)
            encoded_sequence[i, index] = 1

    return encoded_sequence

# Apply one-hot encoding to all sequences in the 'sequence' column
df['sequence_encoded'] = df['sequence'].apply(one_hot_encode)
```
